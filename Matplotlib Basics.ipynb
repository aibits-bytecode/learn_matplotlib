#### Legends

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Max')
plt.plot(days,min_t, label='Min')
plt.plot(days,avg_t, label='Avg')
plt.legend(loc='upper right', shadow=True)
```

#### Bar Charts

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

#### How to deal with string values

```python
plt.bar(company, revenue)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue, label='revenue')
plt.ylabel('revenue(bln)')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue, label='revenue')
plt.bar(xpos,profit, label='profit')
plt.ylabel('revenue(bln)')
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.4, label='revenue')
plt.bar(xpos+0.2,profit, width=0.4,label='profit')
plt.ylabel('revenue(bln)')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.barh(xpos-0.2,revenue,height=0.4, label='revenue')
plt.barh(xpos+0.2,profit,height=0.4, label='profit')
plt.xlabel('revenue(bln)')
plt.legend()
```

### Histograms üòá

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['b','g'], 
         label=['men','women'])
plt.xlabel('Suger range')
plt.ylabel('Number of patients')
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['b','g'], 
         label=['men','women'], orientation='horizontal')
plt.ylabel('Suger range')
plt.xlabel('Number of patients')
plt.legend()
```

### Pie chart üòá

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels)
#plt.legend()
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=0.9)
#plt.legend()
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.2, autopct='%0.2f%%')
#plt.legend()
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.2, autopct='%0.2f%%', explode=[0,0,0.2,0.2,0])
#plt.legend()
plt.show()
 
```

#### Save chart file

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.2, autopct='%0.2f%%', explode=[0,0,0.2,0.2,0])
#plt.legend()
plt.savefig('data/piechart.png', bbox_inches='tight', transparent=True)
```
